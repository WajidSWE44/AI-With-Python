1. Write a program that takes salary as input. Using conditional statements, calculate the final tax based on the following rules: (Marks 3)
•If the salary is less than 30,000 → Tax rate is 5%
•If the salary is between 30,000 and 70,000 → Tax rate is 15%
•If the salary is greater than 70,000 → Tax rate is 25%

```python
salary = float(input("Enter your salary: "))

if salary < 30000:
    tax = salary * 0.05
elif salary >= 30000 and salary <= 70000:
    tax = salary * 0.15
else:
    tax = salary * 0.25

print("Your tax is:", tax)
```

2. Given a list of words: words = ["apple", "banana", "kiwi", "cherry", "mango"] 
Create a dictionary that maps each word to its corresponding length.
Example Output: ({"apple": 5, "banana": 6, "kiwi": 4, "cherry": 6, "mango": 5})

```python
words = ["apple", "banana", "kiwi", "cherry", "mango"]

word_length = {}

for word in words:
    word_length[word] = len(word)

print(word_length)
```

```python
Create a Chat System using Object-Oriented Programming (OOP) concepts.
You need to create the following classes:
•User
•Message
•ChatRoom
The system should implement the following functionalities:
•Sending messages
•Viewing chat history
•User joining and leaving the chat room
```

```python
class Message:
    def __init__(self, sender, content):
        self.sender = sender
        self.content = content

    def display(self):
        print(f"{self.sender}: {self.content}")

class User:
    def __init__(self, name):
        self.name = name
    def send_message(self, room, text):
        new_msg = Message(self.name, text)
        room.add_message(new_msg)

class ChatRoom:
    def __init__(self, room_name):
        self.room_name = room_name
        self.users = []
        self.history = []

    def join(self, user):
        if user not in self.users:
            self.users.append(user)
            print(f"--- {user.name} joined {self.room_name} ---")

    def leave(self, user):
        if user in self.users:
            self.users.remove(user)
            print(f"--- {user.name} left {self.room_name} ---")

    def add_message(self, message):
        self.history.append(message)

    def show_history(self):
        print(f"\n--- {self.room_name} History ---")
        for msg in self.history:
            msg.display()
        print("\n")

room = ChatRoom("Python Enthusiasts")
user_ali = User("Ali")
user_ahmed = User("Ahmed")

room.join(user_ali)
room.join(user_ahmed)

user_ali.send_message(room, "Hello everyone!")
user_ahmed.send_message(room, "Hey Ali, how's the coding going?")

room.show_history()
room.leave(user_ali)
```
