#### Task 3 : Data Cleaning:                                                                                                         
• Identify missing values  
• Handle missing values appropriately  
• Check and remove duplicate records if any 

```python
df.isnull().sum()
```

 Fill numeric columns with median

```python
for col in df.select_dtypes(include=[np.number]).columns:
    df[col] = df[col].fillna(df[col].median())
```

Fill categorical columns with mode

```python
for col in df.select_dtypes(include=["str"]).columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print("\nMissing Values After Handling:")
print(df.isnull().sum())
```

now Remove duplicates

```python
df.drop_duplicates(inplace=True)
print("\nShape After Removing Duplicates:")
print(df.shape)
```

#### Task 4 : Feature Selection:                                                                                                         
• Identify relevant features for analysis  
• Remove unnecessary columns such as identifiers (e.g., Id) 

```python
df.drop(columns=["Id"], inplace=True)
print("\nColumns After Removing Id:")
print(df.columns)
```

#### Task 5: Data Preprocessing:                                                                                                       
• Convert categorical variables into numerical form (encoding)  
• Preparing the dataset for machine learning and display it’s preview in the output  

```python
le = LabelEncoder()
for col in df.select_dtypes(include=["str"]).columns:
    df[col] = le.fit_transform(df[col])
```
